```python
import os
import subprocess

from student_performance_xai.config import PATH_TO_RAW_DATA


PATH_TO_KAGGLE = "/Users/jessicahassibi/PycharmProjects/student-performance-xai/venv/bin/kaggle"
DATA = "larsen0966/student-performance-data-set"


def download_dataset():
    os.makedirs(PATH_TO_RAW_DATA, exist_ok=True)

    command = [
        PATH_TO_KAGGLE,
        "datasets",
        "download",
        "-d", DATA,
        "-p", PATH_TO_RAW_DATA,
        "--unzip"
    ]

    subprocess.run(command, check=True)
    print("Dataset successfully downloaded!")

if __name__ == "__main__":
    download_dataset()
```
